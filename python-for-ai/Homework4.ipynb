2. Write a class definition named Book. The Book class should have data attributes for a book’s
title, the author’s name, and the publisher’s name. The class should also have the following:


An _ _init_ _ method for the class. The method should accept an argument for each of
the data attributes.


```python
class Book:
    def __init__(self, title, author, publisher):
        self.title = title
        self.author = author
        self.publisher = publisher

    # Accessor methods (getters)
    def get_title(self):
        return self.title

    def get_author(self):
        return self.author

    def get_publisher(self):
        return self.publisher

    # Mutator methods (setters)
    def set_title(self, title):
        self.title = title

    def set_author(self, author):
        self.author = author

    def set_publisher(self, publisher):
        self.publisher = publisher

    # String representation of the object
    def __str__(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nPublisher: {self.publisher}"

# Example usage
book = Book("The Great Gatsby", "F. Scott Fitzgerald", "Scribner")

# Accessing attributes using accessor methods
print("Title: ", book.get_title())
print("Author: ", book.get_author())
print("Publisher: ", book.get_publisher())

# Modifying attributes using mutator methods
book.set_title("New Title")
book.set_author("New Author")
book.set_publisher("New Publisher")

# Displaying the book object's state using the __str__ method
print("\nBook Information:\n", book)
```

1. Polymorphism allows you to write methods in a subclass that have the same name as methods in the superclass.


True

```python
class Toy:
  def make_sound(self):
    pass

class Car(Toy):
  def make_sound(self):
    return "vroom vroom!"

class Doll(Toy):
  def make_sound(self):
    return "Hello, I'm a doll!"

class Robot(Toy):
  def make_sound(self):
    return "Beep boop!"

toy1 = Car()
toy2 = Doll()
toy3 = Robot()

toys = [toy1, toy2, toy3]

for toy in toys:
  print(toy.make_sound())
```

2. It is not possible to call a superclass’s _ _init_ _ method from a subclass’s __init__ method.

False

```python
class BigBlock:
  def __init__(self):
    self.size = "Big"

class SmallBlock(BigBlock):
  def __init__(self):
    super().__init__() # This line calls the BigBlock's __init__ method
    self.size = "Small"

small = SmallBlock()
big = BigBlock()
print(small.size)
print(big.size)





```

1. **Employee and ProductionWorker Classes**

Write an Employee class that keeps data attributes for the following pieces of information:
- Employee name
- Employee number


```python
class Employee:
    def __init__(self, name, emp_number):
        self.name = name
        self.emp_number = emp_number

    def display_employee_info(self):
        print(f"Employee name: {self.name}")
        print(f"Employee number: {self.emp_number}")

# Creating an Employee object
employee = Employee("Jane Doe", "A12456")

# Define the ProductionWorker subclass
class ProductionWorker(Employee):
    def __init__(self, name, emp_number, shift_number, hourly_pay_rate):
        # Call the superclass's __init__ method to initialize name and emp_number
        super().__init__(name, emp_number)
        # Initialize the additional attributes for ProductionWorker
        self.shift_number = shift_number
        self.hourly_pay_rate = hourly_pay_rate

    def display_production_worker_info(self):
        self.display_employee_info()
        print(f"Shift: {self.shift_number}")
        print(f"Hourly Pay Rate: ${self.hourly_pay_rate:.2f}")

def main():
    name = input("Enter employee name: ")
    emp_number = input("Enter employee number: ")
    shift_number = int(input("Enter shift number (1 for day shift, 2 for night shift): "))
    hourly_pay_rate = float(input("Enter hourly pay rate: "))

    production_worker = ProductionWorker(name, emp_number, shift_number, hourly_pay_rate)

    print("\nProduction Worker Information:")
    production_worker.display_production_worker_info()

if __name__ == "__main__":
    main()
```

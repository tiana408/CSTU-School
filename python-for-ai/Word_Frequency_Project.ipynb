```python
import PyPDF2

class WordAnalysis:

    def __init__(self):
        self.word_freq = {}
        self.filtered_words = set()

    def ExtractWords(self, text):
        # splitting the words by spaces and punctuations
        words = [word.lower().strip('.,!?:;"\'()[]{}') for word in text.split()]

        # counting the frequency of words
        for word in words:
            if word not in self.filtered_words:
                self.word_freq[word] = self.word_freq.get(word, 0) + 1

    def AddRemovedWords(self, word):
        self.filtered_words.add(word)

    def DelRemovedWords(self, word):
        if word in self.filtered_words:
            self.filtered_words.remove(word)

    def Show(self, freq):
        # filtering words by the given frequency and sorting them
        words_sorted = sorted(self.word_freq.items(), key=lambda x: x[1], reverse=True)

        for word, count in words_sorted:
            if count > freq:
                print(f"{word:15} -- frequency: {count}")

    def __str__(self):
        words_sorted = sorted(self.word_freq.items(), key=lambda x: x[1], reverse=True)
        result = []

        count = 0
        line = ""
        for word, frequency in words_sorted:
            line += f"{word:15} {frequency:3}"
            count += 1
            if count == 4:
                result.append(line)
                line = ""
                count = 0

        if line:  # to add any remaining words not making up to 4
            result.append(line)

        return "\n".join(result)
```

```python
pdfFileObj = open('SampleCh7.pdf', 'rb')
pdfReader = PyPDF2.PdfReader(pdfFileObj)

# Use len(reader.pages) instead of reader.numPages
pages = len(pdfReader.pages)

wordAnalysis = WordAnalysis()

wordAnalysis.AddRemovedWords("and")
wordAnalysis.AddRemovedWords("a")
wordAnalysis.AddRemovedWords("the")

for i in range(pages):  # iterate through all pages
    pageObj = pdfReader.pages[i]
    # Use the new method extract_text instead of extractText
    wordAnalysis.ExtractWords(pageObj.extract_text())

wordAnalysis.Show(10)
print(wordAnalysis)

pdfFileObj.close()
```

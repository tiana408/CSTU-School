```python
def division(x, y):
  try:
    x = int(x)
    y = int(y)
    result = x / y
    return result
  except ZeroDivisionError as e:
    print("Error: Division by zero")
  except ValueError as e:
    print("Error Invalid input value")
  except Exception as e:
    print(f"An unknown error occurred: {e}")

a = input("Enter number a: ")
b = input("Enter number b: ")

result = division(a, b)

if result is not None:
  print(f"Result of division: {result}")
```

```python
def type_check(correct_type):
  def decorator(func):
    def wrapper(arg):
      if isinstance(arg, correct_type):
        return func(arg)
      else:
        print("Bad Type")
    return wrapper
  return decorator

@type_check(int)
def times2(num):
  return num * 2

print(times2(2))
times2('Not A Number')

@type_check(str)
def first_letter(word):
  return word[0]

print(first_letter("Hello World"))
first_letter(['Not', 'A', 'String'])
```

Define a function and use the decorator to print the function name.


```python
# Decorator that takes and prints the name of a func.
def decorator_showname(myfunc):
    def wrapper_func(*args, **kwargs):
        print("I am going to exectue: ", myfunc.__name__)
        return myfunc(*args, **kwargs)
    return wrapper_func

#define a function and apply the decorator
@decorator_showname
def greet(name):
  return f"Hello, {name}!"

#call the decorated function
result = greet("Alice")

#output will include the function name and its result
print(result)
```

5. Write a simple Python function that computes the square of a number. Do not use any TensorFlow operations.
Now, annotate this function with @tf.function.
Compare the outputs of the function before and after the transformation for several inputs. Are they the same?

(Hint:  @tf.function only speed up the computation)


```python
import tensorflow as tf

#define a simple python function to compute the square of a number
def square(x):
  return x * x

#annotate the function with tf.function
@tf.function
def tf_square(x):
  return x * x

#compare the outputs of the function before and after the transformation
inputs = [2.0, 3.0, 4.0]

for x in inputs:
  result1 = square(x)
  result2 = tf_square(x)

  print(f"Input: {x}")
  print(f"Python function result: {result1}")
  print(f"tf.function result: {result2}")
  print()
```

6. Exploring Speedup:

Create a function that performs a large number of matrix multiplications using TensorFlow operations. Time how long it takes to execute this function without the @tf.function annotation.
Now, annotate the function with @tf.function and time it again. Is there a noticeable difference in speed?

Sample Code:
import time

matrix = tf.random.normal([1000, 1000])

def matrix_multiplication():
    return tf.linalg.matmul(matrix, matrix)

@tf.function
def tf_matrix_multiplication():
    return tf.linalg.matmul(matrix, matrix)

start = time.time()
matrix_multiplication()
print("Without @tf.function:", time.time() - start)

start = time.time()
tf_matrix_multiplication()
print("With @tf.function:", time.time() - start)


```python
import tensorflow as tf
import time

matrix = tf.random.normal([1000, 1000])

def matrix_multiplication():
  for _ in range(1000): #perform 1000 matrix multiplications
     result = tf.linalg.matmul(matrix, matrix)
  return result

@tf.function
def tf_matrix_multiplication():
  for _ in range(1000): #perform 1000 matrix multiplications
     result = tf.linalg.matmul(matrix, matrix)
  return result

start = time.time()
matrix_multiplication()
print("Without @tf.function:", time.time() - start)

start = time.time()
tf_matrix_multiplication()
print("With @tf.function:", time.time() - start)
```

```python
def logger(cls):
  class Wrapped(cls):
    def __init__(self, *args, **kwargs):
      print(f"Creating an instance of {cls.__name__}")
      super().__init__(*args, **kwargs)

  return Wrapped

@logger
class MyClass:
  def __init__(self, name):
    self.name = name

my_obj = MyClass("Alice")
```

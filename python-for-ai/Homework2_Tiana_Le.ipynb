5. Write code that reads the file “text.txt”, and count how many words it has, and save the word frequency in a dictionary.

```python

file = open("text.txt", "r")
text = file.read()
file.close()
words = text.split()
word_freq = {}

for word in words:
    word = word.strip(".,!?()[]{}\"'").lower()
    if word:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

for word, freq in word_freq.items():
    print(f"{word}: {freq}")

output_file.close()
```

9. Import the dirtydata.csv dataset. After import the dataset, replace all the missing values with max value of that column.

```python
import pandas as pd

df = pd.read_csv('dirtydata.csv')

for column in df.columns:
  df[column] = pd.to_numeric(df[column], errors='coerce')

  max_value = df[column].max()
  df[column].fillna(max_value, inplace=True)

df.to_csv('cleaneddata.csv')


df.head()
```

```python
import pandas as pd

df = pd.read_csv('user_data.csv')

missing_data = df.isna()

missing_counts = missing_data.sum()

print("Missing data counts: ")
print(missing_counts)

df_cleaned = df.dropna()

df_cleaned.to_csv('cleaned_user_data.csv')
```

Check to see how many rows were dropped due to missing data

```python
import pandas as pd

df = pd.read_csv('user_data.csv')

original_rows = len(df)

df_cleaned = df.dropna()

cleaned_rows = len(df_cleaned)

rows_dropped = original_rows - cleaned_rows

print(f"Number of rows dropped due to missing data: {rows_dropped}")

df_cleaned.to_csv('cleaned_user_data.csv', index=False)
```

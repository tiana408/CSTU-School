1. Some problems can be solved through recursion only. True or False? Find an example of a problem that can be solved by recursion and loops.

False.

```python
def sum_recursive(n):
    if n == 0:
        return 0
    else:
        return n + sum_recursive(n - 1)

# Example usage:
result_recursive = sum_recursive(5)
print("Sum using recursion:", result_recursive)
```

```python
def sum_iterative(n):
    result = 0
    for i in range(1, n + 1):
        result += i
    return result

# Example usage:
result_iterative = sum_iterative(5)
print("Sum using a loop:", result_iterative)

```

2. Write a recursive function that accepts an integer argument, n. The function should display n
lines of asterisks on the screen, with the first line showing 1 asterisk, the second line showing 2
asterisks, up to the nth line which shows n asterisks. The output should look like this:


```python


def display_asterisks(n):
  if n > 0:
    display_asterisks(n - 1)
    print('*' * n)

n = int(input("Enter the number of lines: "))
display_asterisks(n)
```

3. Largest List Item

Design a function that accepts a list as an argument and returns the largest value in the list. The
function should use recursion to find the largest item.

```python
def find_largest_value(my_list):
  if len(my_list) == 1:
    return my_list[0]

  mid = len(my_list) // 2
  left_half = my_list[:mid]
  right_half = my_list[mid:]

  left_largest = find_largest_value(left_half)
  right_largest = find_largest_value(right_half)

  return max(left_largest, right_largest)

my_list = [4, 2, 9, 7, 5, 11, 3]
largest_value = find_largest_value(my_list)
print(largest_value)


```

4. Sum of Numbers

Design a function that accepts an integer argument and returns the sum of all the integers from 1 up to the number passed as an argument. For example, if 50 is passed as an argument, the
function will return the sum of 1, 2, 3, 4, . . . 50. Use recursion to calculate the sum.


```python
def sum_recursive(n):
    if n == 1:
        return 1
    else:
        return n + sum_recursive(n - 1)

# Get the value of n from user input
n = int(input("Enter a positive integer: "))

result_recursive = sum_recursive(n)
print("Sum using recursion:", result_recursive)
```

6. Write a Python code to remove all rows containing Nan values using the Bitwise NOT operator and np.isnan() function.

```python
import numpy as np

data = np.array([[1, 2, 3],
                 [4, np.nan, 6],
                 [7, 8, np.nan],
                 [np.nan, 10, 11]])

rows_with_nan = np.isnan(data).any(axis=1)

data_without_nan = data[~rows_with_nan]

print("Original Data:")
print(data)

print("\nData without NaN rows:")
print(data_without_nan)
```

**Ackermann’s Function  (Optional)**

Ackermann’s Function is a recursive mathematical algorithm that can be used to test how well a
system optimizes its performance of recursion. Design a function ackermann(m, n), which
solves Ackermann’s function. Use the following logic in your function:

If m = 0 then return n + 1
If n = 0 then return ackermann(m – 1, 1)
Otherwise, return ackermann(m – 1, ackermann(m, n – 1))

Once you’ve designed your function, test it by calling it with small values for m and n.


```python
def ackermann(m,n):
  if m == 0:
    return n + 1
  elif n == 0:
    return ackermann(m - 1, 1)
  else:
    return ackermann(m - 1, ackermann(m, n - 1))


result = ackermann(1,2)
print("Ackerman(1,2) =", result)
```

Plot red for label 1 (y=0), blue for label 0 (y=1)

````python
def plot_points(X, y):
    admitted = X[np.argwhere(y==1)]
    rejected = X[np.argwhere(y==0)]
    plt.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'red', edgecolor = 'k')
    plt.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'blue', edgecolor = 'k')
````

#### Plot a random separator line

````python
def display(slope, yinter, color='g--'):
    plt.xlim(-0.05,1.05)
    plt.ylim(-0.05,1.05)
    x = np.arange(-10, 10, 0.1)
    plt.plot(x, slope*x + yinter, color)

display(-1,1.5,"r--")
plot_points(X,y)
plt.show()
````

#### Plot separate line using WX + b model
$
\begin{align}
WX + b = 0\\
w_1x_1 + w_2x_2 + b = 0\\
x_2 = -\frac{w_1}{w_2} x_1 - \frac{b}{w_2}
\end{align}
$

````python
def plot_model(w1, w2, b, color='g--'):
    slope = -w1/w2
    yinter = -b/w2
    print(f'slope={slope}, yinter={yinter}')
    display(slope, yinter, color)

w1 = 1
w2 = 1
b = -1.5
plot_model(w1, w2, b,"r--")

r = 0.1      # try 0.1 and 0.01
# A missclassified point (0.8, 0.3) as 0 (blue) but should be 1 (red)
w1 += r*0.8
w2 += r*0.3
#w1 += r*0.7
#w2 += r*0.6
b += r
plot_model(w1, w2, b, "b--")
plot_points(X,y)
plt.show()
````

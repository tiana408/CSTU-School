```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

### Create a matrix x

```python
mu = [0,0]
sigma = [[0.6,0.2],[0.2,0.2]]
n = 1000
x = np.random.multivariate_normal(mu, sigma, n).T
```

### Get its covariance matrix

```python
A=np.cov(x)
```

```python
x
```

```python
A
```

### Get eigenvalues and eigenvectors from covariance matrix

```python
eigenvals, eigenvecs = np.linalg.eig(A)
```

```python
eigenvals
```

```python
eigenvecs
```

### Eigenvector corresponding to the largest eigenvalue shold be the 1st to be plotted as new axle system. Plot line to reflect gradiance of eigenvecs (0,0.92222334) and (0, 0.38665762) as first eigen axles, because 1st eigenval is abs(0.65340822) > 2nd one, abs(-0.38665762); Plot line to reflect (0, -0.38665762), (0.92222334) as second eigen axles


```python
plt.scatter(x[0,:], x[1,:], alpha=0.2)
plt.plot([0, 3*eigenvals[0]*eigenvecs[0][0]], [0, 3*eigenvals[0]*eigenvecs[1][0]], 'r-', lw=2) #larger eigenvecs
plt.plot([0, 3*eigenvals[1]*eigenvecs[0][1]], [0, 3*eigenvals[1]*eigenvecs[1][1]], 'r-', lw=2) #smaller eigenvecs
plt.plot([0, -3*eigenvals[0]*eigenvecs[0][0]], [0, -3*eigenvals[0]*eigenvecs[1][0]], 'r-', lw=2) #larger eigenvecs
plt.plot([0, -3*eigenvals[1]*eigenvecs[0][1]], [0, -3*eigenvals[1]*eigenvecs[1][1]], 'r-', lw=2) #smaller eigenvecs
```

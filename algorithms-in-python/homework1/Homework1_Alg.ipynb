20. Valid Parentheses

O(n)

```python
def isValid(s):
  stack = []
  pairs = {
     '(': ')',
     '{': '}',
     '[': ']'
  }
  for bracket in s:
    if bracket in pairs:
      stack.append(bracket)
    elif len(stack) == 0 or \
        bracket != pairs[stack.pop()]:
      return False
  return len(stack) == 0



print(isValid("()"))
print(isValid("()[]{}"))
print(isValid("(}"))
```

58. Length of Last Word

O(n)

```python
def lengthOfLastWord(s):
  stripped_s = s.strip()

  words = stripped_s.split()

  return len(words[-1])

print(lengthOfLastWord("Hello World"))
print(lengthOfLastWord("   fly me   to   the moon  "))
print(lengthOfLastWord("luffy is still a joyboy"))
```
